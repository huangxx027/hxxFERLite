## emo_list数组记录每一帧的表情

```python
emo_list = df1['emotion'].tolist()
```

对课堂情绪进行分类

```python
#将下标【所在帧的序号】存入各类
#基础类型
sad = []
anger = []
disgust = []
fear = []
contempt = []
happy = []
surprised = []
neutral = []

#积极与消极
positive = []
negative = []

#倾听理解疑惑抵抗
listen = []
understanding = []
confusing = []
resistant = []

for k in range(len(emo_list)):
    if(emo_list[k] == '伤心'):
        sad.append(k)
        confusing.append(k)
        negative.append(k)
        
    if(emo_list[k] == '发怒'):
        anger.append(k)
        confusing.append(k)
        negative.append(k)
        
    if(emo_list[k] == '恐惧'):
        fear.append(k)
        resistant.append(k)
        negative.append(k)
        
    if(emo_list[k] == '厌恶'):
        disgust.append(k)
        resistant.append(k)
        negative.append(k)
        
    if(emo_list[k] == '开心'):
        happy.append(k)
        understanding.append(k)
        positive.append(k)
        
    if(emo_list[k] == '惊讶'):
        surprised.append(k)
        understanding.append(k)
        positive.append(k)
        
    if(emo_list[k] == '中性'):
        neutral.append(k)
        listen.append(k)
        positive.append(k)
        
    if(emo_list[k] == '不屑'):
        contempt.append(k)
        resistant.append(k)
        negative.append(k)
        
print(positive)
print(negative)
```

学生课堂活跃度分析

```python
#积极/消极表情占比
positive_percentage = len(positive)/len(emo_list)
negative_percentage = len(negative)/len(emo_list)
print(positive_percentage)
print(negative_percentage)
```

```python
print("该学生的课堂积极度指数为"+str(positive_percentage)+"；课堂消极度指数为"+str(negative_percentage))
if(positive_percentage>negative_percentage):
    print("课堂活跃度良好！")
else:
    print("课堂活跃度一般。")
```

课堂情绪变化分析

```python
#定义帧数转时间函数，以fps=25为例，返回视频时间
def frametoSec(frame_num):
    sec = frame_num/25
    return sec
```

```python
#定义记录每一帧为积极/消极的数组 EmotionEveryFrame
EmotionEveryFrame = []
for i in range(len(emo_list)):
    if(emo_list[i]=='开心'or emo_list[i]=='中性' or emo_list[i]=='惊讶'):
        EmotionEveryFrame.append(1)
    if(emo_list[i] == '恐惧'or emo_list[i] == '不屑'or emo_list[i] == '厌恶'or emo_list[i] == '发怒'or emo_list[i] == '伤心'):
        EmotionEveryFrame.append(0)
    if(emo_list[i]=='无'):
        EmotionEveryFrame.append(-1) 
print(EmotionEveryFrame)
```

输出听懂、听不懂、听不懂很久的时间（警告）

```python
#switch_timing 存变换的时间
switch_timing = []
#last_emo_type存最后一次上一个类型出现的下标
#emo_type存变换后的类型的下标
last_emo_type = EmotionEveryFrame[0]
for emo_type in range(len(EmotionEveryFrame)):
     if(EmotionEveryFrame[emo_type] != EmotionEveryFrame[last_emo_type]):
            if(EmotionEveryFrame[emo_type]==-1):
                print("学生在" + str(frametoSec(last_emo_type)) + "到" + str(frametoSec(emo_type)) + "没有看向老师")
            if(EmotionEveryFrame[emo_type]==0):
                print("学生在"+str(frametoSec(last_emo_type))+"到"+str(frametoSec(emo_type))+"听懂了")
            if(EmotionEveryFrame[emo_type]==1):
                #这里是大于3秒就提示太长，可供自行修改
                if((frametoSec(emo_type) - frametoSec(last_emo_type)) > 3):
                   print("该学生在"+str(frametoSec(last_emo_type))+"到"+str(frametoSec(emo_type))+"一直没听懂（长时间处于消极状态），教职工应重点关注该学生。")
                else:
                    print("学生在"+str(frametoSec(last_emo_type))+"到"+str(frametoSec(emo_type))+"听不懂")
            switch_timing.append(frametoSec(emo_type) - frametoSec(last_emo_type))
            last_emo_type = emo_type

#print(switch_timing)
```

输出为四种情况：
（1）学生在这些时间段听懂了
（2）学生在这些时间段没听懂
（3）学生在这段时间一直没听懂（一直处于消极状态），教职工应重点关注该学生。
（4）学生在这些时间段没有看向老师
